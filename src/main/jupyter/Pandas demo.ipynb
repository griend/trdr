# Pandas demo

Load Python modules and setup the connection with the SQLite database.

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd
import sqlite3

db_filename = '/Users/cees/var/trader/db/trader.db'

connection = sqlite3.connect(db_filename)
connection.row_factory = sqlite3.Row
cursor = connection.cursor()
```

Create a Pandas Dataframe by querying the SQLite database.

```python
df = pd.read_sql('SELECT * FROM markets',
                 connection,
                 index_col='id', 
                 parse_dates={'created_on': 's', 'updated_on': 's'})

df.head()
```

A silly example of a plot, because the dataframe only contains metadata.

```python
df.plot()
plt.title('Test')
```
